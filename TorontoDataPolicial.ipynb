```python
crime_data.drop(['Index_','event_unique_id', 'X', 'Y', 'ucr_code', 'ucr_ext', 'FID', 'Neighbourhood'], axis=1, inplace=True)
```

```python
crime_data.head()
```

```python
occurrencedate_list = crime_data.occurrencedate.str.split('T').tolist()
reporteddate_list = crime_data.reporteddate.str.split('T').tolist()

for i in range(len(occurrencedate_list)):
  occurrencedate_list[i] = occurrencedate_list[i][0]

for i in range(len(reporteddate_list)):
  reporteddate_list[i] = reporteddate_list[i][0]

  crime_data['occurrencedate'] = occurrencedate_list
  crime_data['reporteddate'] = reporteddate_list

  crime_data.occurrencedate = pd.to_datetime(crime_data.occurrencedate)
  crime_data.reporteddate = pd.to_datetime(crime_data.reporteddate)
```

```python
crime_data.drop(['reportedyear','','','','','','',''],axis=1, inplance=True)
```

```python
crime_data.dtypes
```

```python
crime_data.premisetype = crime_data.premisetype.astype('category', error='raise')
crime_data.offence = crime_data.offence.astype('category', error='raise')
crime_data.MCI = crime_data.MCI.astype('category', error='raise')
crime_data.reporteddayofweek = crime_data.reporteddayofweek.astype('category', error='raise')
crime_data.reporteddayofweek = crime_data.reporteddayofweek.str.rstrip().astype('category', error='raise')
crime_data.Division = crime_data.Division.astype('category', error='raise')
```

```python
def doKMeans(data, n_clusters):
  fig = plt.figure()
  ax = fig.add_subplot(111)
  ax.scatter(data.Long. data.Lat, marker='.', c='blue')

  data = data.loc[:, ['Long, Lat']]

  model = KMeans(n_clusters=n_clusters)
  model.fit(data)

  centroids = model.cluster_centers_
  print(centroids)

  ax.scatter(centroids[:,0], centroids[:,1], marker='x', c='red')
  plt.show()

```

```python
break_enter = crime_data[(crime_data.MCI == 'Break and Ener') & (crime_data.premisetype == 'Apartment')]

break_enter.reset_index(inplance=True, drop=True)
```
